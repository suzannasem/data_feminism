```python
import pandas as pd 
import numpy as np
```

```python
df_example = pd.read_csv('audible_uncleaned.csv')
df_example
```

```python
## AUTHOR 

#df_example['author'].str.lstrip('Writtenby:') # first I checked to see if this gave us the column we wanted

df_example['author'] = df_example['author'].str.lstrip('Writtenby:') ## now we can overwrite the original column with the above code

df_example
```

```python
import matplotlib.pyplot as plt  ## importing a package to create a visualization 

valueCounts = df_example['author'].head(100).value_counts()  ## counting up the number of times an author appears in the first 100 lines 

valueCounts.plot(kind='bar')
```

```python
## NARRATOR

df_example['narrator'] = df_example['narrator'].str.lstrip('Narratedby:') ## now we can overwrite the original column with the above code

df_example
```

```python
## TIME
def convertTime(timeString):
    totalMins = 0
    splitString = timeString.split(' and ') ## split on the and statement 
    for item in splitString:
        if 'hrs' in item:
            hours = int(item.split(' ')[0])
            totalMins += (60*hours)
        elif 'mins' in item:
            mins = int(item.split(' ')[0])
            totalMins += mins
    return totalMins


convertTime("2 hrs and 10 mins")
#df_example['time'] = ...
#df_example
```

```python
df_example['time'] = df_example['time'].apply(convertTime)  ## how to apply a function to a column pandas: https://stackoverflow.com/questions/34962104/how-can-i-use-the-apply-function-for-a-single-column

df_example 
```

```python
## RELEASE DATE

df_example['releasedate'] = pd.to_datetime(df_example["releasedate"], infer_datetime_format=True, format='mixed').dt.strftime('%Y-%m-%d')
df_example
```

```python
## STARS

notRated = df_example[df_example['stars'] == 'Not rated yet']  ## select rows where nothing is rated 
rated = df_example[df_example['stars'] != 'Not rated yet']  ## select all the rows where there is a rating

rated
```

```python
rated[['stars', 'numRatings']] = rated['stars'].str.split('stars', expand = True)
rated
```

```python
rated['stars'] = rated['stars'].str.split(' out of ').str[0].astype(float)
rated 
```

```python
rated['numRatings'] = rated['numRatings'].str.split(' ').str[0].str.replace(',', '').astype(int)
rated
```

```python
notRated['stars'] = np.nan 
notRated['numRatings'] = 0

notRated
```

```python
df_cleaned = pd.concat([rated, notRated]) ## put the tables back together 

df_cleaned
```
